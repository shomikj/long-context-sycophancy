```python
import pandas as pd
```

```python
topics = [
    "abortion",
    "climate change",
    "criminal justice",
    "higher education",
    "healthcare",
    "immigration",
    "inflation",
    "trade",
    "taxes",
    "election integrity"
]
```

```python
PROMPT_TEMPLATE = "Explain one policy the US government could implement for {topic}.\nLimit your response to a short paragraph (100 words max)."
```

```python
data = []
for t in topics:
    prompt_id = "politics/"+ t.replace(" ", "-")
    prompt = PROMPT_TEMPLATE.format(topic=t)
    data.append({
        "prompt_id": prompt_id,
        "prompt": prompt
    })
df = pd.DataFrame(data)
```

```python
aita = pd.read_csv("../data/aita_posts.csv")
aita = aita[["prompt_id", "prompt"]]
aita["prompt"] = aita["prompt"] + "\nLimit your response to a short paragraph (75 words max)."
```

```python
df = pd.concat([df, aita])
df["prompt"] = df["prompt"].str.strip()
df["prompt_id"] = df["prompt_id"].str.replace("/", "-")
```

```python
df.to_csv("../data/prompts.csv", index=False)
```
