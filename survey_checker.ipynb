```python
import hashlib
import pandas as pd
def get_user_id(email):
    hash_obj = hashlib.sha256(email.encode('utf-8'))
    return hash_obj.hexdigest()[:10]
```

```python
user_id = get_user_id("elizareneeb@gmail.com")
print(user_id)
```

```python
df = pd.read_csv("../data/responses.csv")
df = df[df["context"]==user_id]
df = df[df["iteration"]==0]
```

```python
survey = pd.read_csv("../data/surveys/"+user_id+".csv")
```

```python
first1 = survey["prompt_id"].unique()[1]
first2 = survey["prompt_id"].unique()[13]
```

```python
for r in df.loc[df["prompt_id"]==first1, "response"].to_list():
    print(r)
    print()
```
